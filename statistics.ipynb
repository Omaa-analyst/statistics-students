```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
```

```python
stats = pd.read_csv(r"C:\Users\USER\OneDrive\Documents\Statistics_Undergraduate_Programs_Ankara.csv", header=0, sep=",")
stats.head(60)
```

```python
stats.columns = stats.columns.str.strip()

stats['University'] = stats['University'].str.strip().str.title()
stats['Internship Experience'] = stats['Internship Experience'].str.strip().str.capitalize()
stats['Concern Level'] = stats['Concern Level'].astype('category')
stats['Anxiety Level'] = stats['Anxiety Level'].astype('category')

stats.info()
```

```python
msno.bar(stats)
```

```python
print("University counts:\n", stats['University'].value_counts(), "\n")
print("GPA stats:\n", stats['GPA'].describe())
```

```python
plt.figure(figsize=(15,8))
sns.countplot(data=stats, x='University', hue='Concern Level')
plt.title("Concern About AI by University")
plt.ylabel("Number of Students")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=stats, x='Impact of AI on Career Plans', y='GPA')
plt.title("GPA vs Perceived Impact of AI on Career Plans")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
cross = pd.crosstab(stats['Internship Experience'], stats['Concern Level'], normalize='index') * 100
cross
```

```python
cross.plot(kind='bar', stacked=True, colormap='coolwarm', figsize=(8,6))
plt.title("Internship vs Concern About AI")
plt.ylabel("Percentage")
plt.xticks(rotation=0)
plt.legend(title='Concern Level')
plt.show()
```

```python
pivot = pd.crosstab(index=[stats['University'], stats['Internship Experience']],
                    columns=stats['Impact of AI on Career Plans'],
                    normalize='index') * 100

plt.figure(figsize=(10,6))
sns.heatmap(pivot, annot=True, fmt=".1f", cmap="YlGnBu")
plt.title("Career Impact by University and Internship Experience")
plt.show()
```

```python
from scipy.stats import chi2_contingency  

# Create a contingency table (frequency table)
contingency_table = pd.crosstab(stats['University'], stats['Concern Level'])

# Perform Chi-square test of independence
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Display results
print("Chi-square Statistic:", chi2)
print("Degrees of Freedom:", dof)
print("P-value:", p)

# Interpret result
if p < 0.05:
    print(" There is a significant association between University and Concern Level.")
else:
    print(" No significant association between University and Concern Level.")
```

```python
# Prepare a simplified dataset for pairplot
plot_stats = stats[['GPA', 'University', 'Concern Level', 'Impact of AI on Career Plans', 'Internship Experience']].dropna()

# Optional: Rename long columns for easier plotting
plot_stats.rename(columns={
    'Concern Level': 'Concern',
    'Impact of AI on Career Plans': 'Impact',
    'Internship Experience': 'Internship'
}, inplace=True)

# Convert GPA ranges to numeric midpoints for plotting
def gpa_midpoint(gpa_range):
    try:
        low, high = gpa_range.split('-')
        return (float(low.strip()) + float(high.strip())) / 2
    except Exception:
        return np.nan

plot_stats['GPA_num'] = plot_stats['GPA'].apply(gpa_midpoint)

# Only keep numeric column(s) for pairplot
sns.pairplot(plot_stats, vars=['GPA_num'], hue='University', diag_kind='kde', height=10.5)
plt.suptitle("Pairplot: GPA vs Categorical Features", y=1.02)
plt.show()
```
